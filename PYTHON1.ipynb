```
import pandas as pd
```

```
df=pd.read_csv("/content/Rossmann Stores Data.csv")
```

```
df
```

```
# prompt: linear regression graph between Sales vs Store

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'Sales' and 'Store' are columns in your DataFrame 'df'
X = df['Store'].values.reshape(-1, 1)  # Reshape to a 2D array for sklearn
y = df['Sales'].values

# Create a linear regression model
model = LinearRegression()
model.fit(X, y)

# Predict sales based on the store
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, y_pred, color='red')
plt.xlabel('Store')
plt.ylabel('Sales')
plt.title('Linear Regression: Sales vs Store')
plt.show()
```
